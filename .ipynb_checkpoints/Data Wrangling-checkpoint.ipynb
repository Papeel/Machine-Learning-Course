## Generación aleatoria de número

```python
 import numpy as np
```

```python
#Generar un número aleatorio entero entre 1 y 100
np.random.randint(1,100)
```

```python
#Genera valores entre 0 y 1
np.random.random()
```

```python
def randint_list(n, a, b):
    x=[]
    for i in range(n):
        x.append(np.random.randint(a, b))
        
    return x
```

```python
randint_list(25, 1, 50)
```
